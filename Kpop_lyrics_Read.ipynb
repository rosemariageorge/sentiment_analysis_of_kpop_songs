```python
import numpy as np
import os
import io
import pandas as pd
import re
from os import listdir
from os.path import isfile, join
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
os.getcwd()
data_folder = os.path.join(os.getcwd(),'/content/gdrive/My Drive/PROJECT_THESIS/DATASET/KPOP_LYRICS/LYRICS_CLEANED/')  # path to raw song files
```


Reading the first dataset

```python
file_temp = [os.path.join(data_folder,i) for i in sorted(os.listdir(data_folder)) if os.path.isfile(os.path.join(data_folder,i))]

df = pd.DataFrame()

#for every file in folder, read it and append to a empty dataframe with column filename as 'Title'
for file in file_temp:
    print(file)
    _df = pd.read_csv(file, names =['lyrics'], header=None, index_col=False, sep='\\n', encoding='utf-8')
    _df['Title'] = os.path.split(file)[-1]
    if _df.shape[0] > 6:               # rejecting songs which are under 6 lines
        df = df.append(_df)
```

```python
df_group = df.groupby('Title').agg({'lyrics': lambda x: " ".join(x)}).reset_index()
# Remove .txt extension from the Title column
df_group['Title'] = df_group['Title'].str.replace(r'.txt$', '')
pd.options.display.max_colwidth = 5000
df_group.head()
```

```python
df3_group = df_group.to_csv('lyricscleaned.csv', index=False)
```
